# Hiring Algorithm
This jupyter notebook will focus on the resumes to find out if the person will be hired, based on the stated requirements within the game. These requirements entail: <br>
Must have all of the following:<br>
A bachelor's degree in Accounting or Finance<br>
Proficiency of Microsoft Office<br>

These requirements are tested throughout this notebook, the results added as columns

```python
import pandas as pd
import re 
import math
df = pd.read_csv('resumes2.csv', delimiter= ';')
df.head()
```

```python
df.groupby(['gender']).count()
```

```python
df['gender'] = df['gender'].str.replace(' +$', '')
df.groupby(['gender']).count()
```

```python
df['skills'] = df['skills'].str.replace('[^A-Za-z0-9]+', ' ').str.lower()
df['experience'] = df['experience'].str.replace('[^A-Za-z0-9]+', ' ').str.lower()
df['education'] = df['education'].str.replace('[^A-Za-z0-9]+', ' ').str.lower()
df['motivation'] = df['motivation'].str.replace('[^A-Za-z0-9]+', ' ').str.lower()
df.head()
```

```python
for index,row in df.iterrows():
    for column in df[['skills', 'experience', 'motivation']]:
        no_digits = []
        # Iterate through the string, adding non-numbers to the no_digits list
        if(type(row[column]) == str):
            for i in row[column]:
                if not i.isdigit():
                    no_digits.append(i)
        # Now join all elements of the list with '', 
        # which puts all of the characters together.
            result = ''.join(no_digits)
            df.loc[index,column] = result
        
df
```

```python
for index,row in df.iterrows():
    no_digits = []
    for i in row['education']:
        if not i.isdigit():
                    no_digits.append(i)
        # Now join all elements of the list with '', 
        # which puts all of the characters together.
        result = ''.join(no_digits)
        df.loc[index,'education'] = result
        
df
```

## Gender prediction
Find words that are correlated with a gender. Results are added to femaleConfirmed/maleConfirmed columns

```python
df['femaleConfirmed'] = 0;
df['maleConfirmed'] = 0;
df
```

```python
male_words=["men","mens","boy","boys"]
female_words=["women", "womens", "girl", "girls"]
```

```python
#function to find whole word instead of 'men' in management
def findWord(s, w):
    return (' ' + w + ' ') in (' ' + s + ' ')
```

```python
# Iterate over two given columns only from the dataframe
for column in df[['experience', 'education', 'motivation']]:
    for index,row in df.iterrows():  
        if(pd.notna(row[column])):
        
            for item in female_words:
                female_match = findWord(row[column],item)
                if(female_match != False):
                    df.loc[index,'femaleConfirmed'] = 1
                    
            for item in male_words:
                male_match = findWord(row[column],item)
                if(male_match != False):
                    df.loc[index,'maleConfirmed'] = 1
        
df
```

```python
df['gender'] = df['gender'].factorize()[0]
df = df[pd.notnull(df['motivation'])]
df
```

## Find out if applicant has an appropriate bachelor's degree

```python
df['bachelor'] = 0;
df
```

```python
for column in df[['education']]:
    for index,row in df.iterrows():  
        if(pd.notna(row[column])):
            bachelor = [sentence + '.' for sentence in row[column].split('.') if 'bachelor' in sentence]
            
            bachelor2 = [sentence + '.' for sentence in row[column].split('.') if 'Bsc' in sentence]
            
            bachelor3 = [sentence + '.' for sentence in row[column].split('.') if 'Bsa' in sentence]
   
            if(bachelor):
                wantedBachelor = [sentence + '.' for sentence in str(bachelor).split('.') if 'finance' in sentence]
                wantedBachelor2 = [sentence + '.' for sentence in str(bachelor).split('.') if 'account' in sentence]
                if(len(wantedBachelor)>0):
                    df.loc[index,'bachelor'] = 1
                
                elif(len(wantedBachelor2)>0):
                    df.loc[index,'bachelor'] = 1
                    
            if(bachelor2):
                wantedBachelor = [sentence + '.' for sentence in str(bachelor2).split('.') if 'finance' in sentence]
                wantedBachelor2 = [sentence + '.' for sentence in str(bachelor2).split('.') if 'account' in sentence]
                if(len(wantedBachelor)>0):
                    df.loc[index,'bachelor'] = 1
                
                elif(len(wantedBachelor2)>0):
                    df.loc[index,'bachelor'] = 1
                    
            if(bachelor3):
                wantedBachelor = [sentence + '.' for sentence in str(bachelor3).split('.') if 'finance' in sentence]
                wantedBachelor2 = [sentence + '.' for sentence in str(bachelor3).split('.') if 'account' in sentence]
                if(len(wantedBachelor)>0):
                    df.loc[index,'bachelor'] = 1
                
                elif(len(wantedBachelor2)>0):
                    df.loc[index,'bachelor'] = 1
df
```

## Find out if applicant has sufficient knowledge of Microsoft Office

```python
for column in df[['skills']]:
    for index,row in df.iterrows():  
        if(pd.notna(row[column])):
            microsoft = [sentence + '.' for sentence in row[column].split('.') if 'microsoft office' in sentence]
            if(len(microsoft)>0):
                df.loc[index,'microsoft'] = 1
df
```
