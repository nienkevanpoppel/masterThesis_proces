# 1. Gender decoding algorithm on trainingset

## Preprocessing training dataset

```python
import pandas as pd
import re 
import math
df = pd.read_csv('training_set.csv', delimiter= ',')
df
```

```python
df['gender'] = df['gender'].replace({0:1, 1:0})
df
```

```python
## some column names had spaces in them
df.columns = ['person_name', 'person_gender', 'person_skills', 'person_experience', 'person_education', 'person_qualities', 'femaleConfirmed', 'maleConfirmed']
```

```python
# get all words from qualities en skills columns as seperate columns with counts
df = pd.concat([df,df.person_qualities.str.get_dummies(' ')],axis=1)
df = pd.concat([df,df.person_skills.str.get_dummies(' ')],axis=1)
df
```

```python
#make list of column names to later be used as X value in algorithm
columnList = ['femaleConfirmed','girls', 'people', 'independent', 'player', 'multi', 'information', 'learn', 'drive', 'experience', 'accounting', 'quality', 'efficient', 'well', 'produce', 'provide', 'challenging', 'learning', 'she', 'hardworking', 'her', 'excellence', 'complex', 'mature', 'organized', 'netball', 'interpersonal', 'time', 'basketball', 'soccer', 'football', 'leadership', 'driven', 'business', 'badminton', 'equity', 'analytical', 'his', 'investment', 'proven' , 'keen', 'leader', 'individual', 'sports', 'independently', 'he']
```

## The algorithm

```python
df = df.groupby(level=0, axis=1).sum()
df
```

```python
y = df['person_gender'] #We need to take out the gender as our Y-variable
X = df[columnList] #this slices the dataframe to include all rows I need
X = X.groupby(X.columns, axis=1).sum()
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
train_data_features=X_train
forest = RandomForestClassifier(n_estimators = 100, random_state=1) 
forest = forest.fit( train_data_features, y_train)
forest
```

```python
femConf1 = X_test['femaleConfirmed'].to_numpy()
len(femConf1)
```

## Accuracy of random forest classifier

```python
from sklearn.metrics import accuracy_score
result = forest.predict(X_test)
len(result)
for i, res in enumerate(result):
    if(res == 0 and femConf1[i] == 1):
        result[i] = 1
print(accuracy_score(y_test, result))
```

# 2. Using algorithm on the game dataset

The previously established model will now be used on a smaller dataset consisting of gender neutral names and their details. Gender will be deduced from a persons experience, education and qualities. 

```python
df_game = pd.read_csv('game_set.csv', delimiter= ';')
df_game
```

```python
df_game.columns = ['person_name', 'person_gender', 'person_skills', 'person_experience', 'years_experience', 'person_education', 'person_qualities', 'languages', 'expected_salary']
df_game
```

```python
#make person_gender column into zero's and ones instead of words
df_game['person_gender'] = df_game['person_gender'].factorize()[0]
df_game
```

```python
df_subset = df_game[['person_gender','person_skills', 'person_experience' ,'person_education','person_qualities']]
df_subset['person_skills'] = df_subset['person_skills'].str.replace('[^A-Za-z0-9]+', ' ').str.lower()
df_subset['person_experience'] = df_subset['person_experience'].str.replace('[^A-Za-z0-9]+', ' ').str.lower()
df_subset['person_education'] = df_subset['person_education'].str.replace('[^A-Za-z0-9]+', ' ').str.lower()
df_subset['person_qualities'] = df_subset['person_qualities'].str.replace('[^A-Za-z0-9]+', ' ').str.lower()
df_subset
```

```python
for index,row in df_subset.iterrows():
    for column in df_subset[['person_skills', 'person_experience', 'person_education', 'person_qualities']]:
        no_digits = []
        # Iterate through the string, adding non-numbers to the no_digits list
        if(type(row[column]) == str):
            for i in row[column]:
                if not i.isdigit():
                    no_digits.append(i)
        # Now join all elements of the list with '', 
        # which puts all of the characters together.
            result = ''.join(no_digits)
            df_subset.loc[index,column] = result
        
df_subset
```

## find confirmed

```python
df_subset['femaleConfirmed'] =0
df_subset['maleConfirmed']=0
```

```python
male_words=["men","mens","boy","boys"]
female_words=["women", "womens", "girl", "girls"]
```

```python
#function to find whole word instead of 'men' in management
def findWord(s, w):
    return (' ' + w + ' ') in (' ' + s + ' ')
```

```python
# Iterate over two given columns only from the dataframe
for column in df_subset[['person_experience', 'person_education', 'person_qualities']]:
    for index,row in df_subset.iterrows():  
        if(pd.notna(row[column])):
        
            for item in female_words:
                female_match = findWord(row[column],item)
                if(female_match != False):
                    df_subset.loc[index,'femaleConfirmed'] = 1
                    
            for item in male_words:
                male_match = findWord(row[column],item)
                if(male_match != False):
                    df_subset.loc[index,'maleConfirmed'] = 1
        
df_subset
```

```python
df_subset = pd.concat([df_subset,df_subset.person_qualities.str.get_dummies(' ')],axis=1)
df_subset = pd.concat([df_subset,df_subset.person_skills.str.get_dummies(' ')],axis=1)
df_subset
```

```python
for column in columnList:
    if column not in df_subset.columns:
        df_subset[column]=0
df_subset
```

```python
df_subset = df_subset.groupby(level=0, axis=1).sum()
df_subset['business']
```

```python
X_game = df_subset[columnList] 
# list(X_game)
```

```python
result_game = forest.predict(X_game)
result_game

df_subset['predicted_gender'] = result_game
df_subset['femaleConfirmed']
```

# 3. Apply hiring algorithm
The applicant needs to have/be at least the following:
- 3 years of experience
- A Bachelor's in Accountancy or Finance
- Good at team work

And can only miss one of the following:
- Microsoft office proficiency
- speak at least 2 languagese (bonus points if more)
- Expect a maximum of 3000 as salary  

## Check for correct bachelor

```python
df_game['bachelor'] = 0
df_subset['person_education'] = df_subset['person_education'].str.lower()
for column in df_subset[['person_education']]:
    for index,row in df_subset.iterrows():  
        if(pd.notna(row[column])):
            bachelor = [sentence + '.' for sentence in row[column].split('.') if 'bachelor' in sentence]
            
            bachelor2 = [sentence + '.' for sentence in row[column].split('.') if 'bsc' in sentence]
            
            bachelor3 = [sentence + '.' for sentence in row[column].split('.') if 'bsa' in sentence]
   
            if(bachelor):
                wantedBachelor = [sentence + '.' for sentence in str(bachelor).split('.') if 'finance' in sentence]
                wantedBachelor2 = [sentence + '.' for sentence in str(bachelor).split('.') if 'account' in sentence]
                if(len(wantedBachelor)>0):
                    df_game.loc[index,'bachelor'] = 1
                
                elif(len(wantedBachelor2)>0):
                    df_game.loc[index,'bachelor'] = 1
                    
            if(bachelor2):
                wantedBachelor = [sentence + '.' for sentence in str(bachelor2).split('.') if 'finance' in sentence]
                wantedBachelor2 = [sentence + '.' for sentence in str(bachelor2).split('.') if 'account' in sentence]
                if(len(wantedBachelor)>0):
                    df_game.loc[index,'bachelor'] = 1
                
                elif(len(wantedBachelor2)>0):
                    df_game.loc[index,'bachelor'] = 1
                    
            if(bachelor3):
                wantedBachelor = [sentence + '.' for sentence in str(bachelor3).split('.') if 'finance' in sentence]
                wantedBachelor2 = [sentence + '.' for sentence in str(bachelor3).split('.') if 'account' in sentence]
                if(len(wantedBachelor)>0):
                    df_game.loc[index,'bachelor'] = 1
                
                elif(len(wantedBachelor2)>0):
                    df_game.loc[index,'bachelor'] = 1
df_game
```

## Check for team player qualities

```python
df_game['is_teamplayer'] = 0
for index,row in df_game.iterrows():  
    if(pd.notna(row['person_qualities'])):
        if 'team player' in row['person_qualities']: 
            df_game['is_teamplayer'] =1
        if 'teamplayer' in row['person_qualities']: 
            df_game['is_teamplayer'] =1
        if 'team work' in row['person_qualities']: 
            df_game['is_teamplayer'] =1  
        if 'teamwork' in row['person_qualities']: 
            df_game['is_teamplayer'] =1
```

## Check for microsoft Office skills

```python
df_game['microsoft'] = 0
df_subset['person_skills'] = df_subset['person_skills'].str.lower()
for column in df_subset[['person_skills']]:
    for index,row in df_subset.iterrows():  
        if(pd.notna(row[column])):
            microsoft = [sentence + '.' for sentence in row[column].split('.') if 'microsoft' in sentence]
            if(len(microsoft)>0):
                df_game.loc[index,'microsoft'] = 1
df_game
```

## Converting important values to integers
This is where both algorithms will make calculations if the applicant should be hired. The 'hired' column will represent the fair algorithm, the 'hired_bias' represents the biased algorithm. A 1 meaning they were hired, a 0 meaning they were rejected. The biased algorithm will check if it is a male, and otherwise automatically decline. In order to ensure that the user of the game will be provided with an explanation as to why the user was not hired, these values were saved in the 'missingValues' column.

```python
df_game['expected_salary'] = pd.to_numeric(df_game['expected_salary'], errors='ignore')
# df_game['bachelor'] = pd.to_numeric(df_game['bachelor'], errors='ignore')
# df_game['is_teamplayer'] = pd.to_numeric(df_game['is_teamplayer'], errors='ignore')
```

```python
df_game['hired_bias'] = 0
df_game['hired'] = 0
df_game['missingValues'] = ''
for index,row in df_game.iterrows():
    totalRequired = 0
    totalOptional = 0
    missingvalues = ''
    if(row['years_experience'] >2):
        totalRequired +=1
    else: 
        missingvalues += 'Applicant does not have enough years of relevant experience.'
    if(row['bachelor'] ==1):
        totalRequired +=1
    else:
        missingvalues += 'Applicant does not have a correct bachelor.'
    if(row['is_teamplayer']==1):
        totalRequired +=1
    else:
         missingvalues += 'Applicant is not a team player.'
    if(row['microsoft']==1):
        totalOptional +=1
    else:
         missingvalues += 'Applicant has no microsoft experience.'
    if(row['languages'] >1):
        totalOptional += 1
    else:
         missingvalues += 'Applicant does not speak the minimum of two languages.'
    if(row['expected_salary'] < 3001):
        totalOptional += 1
    else:
         missingvalues += 'Applicant expects a salary of over 3000.'

    if(len(missingvalues) >0):
        df_game.at[index,'missingValues'] = missingvalues
        print(missingvalues)
    else:
        df_game.at[index,'missingValues'] = 'none'
    
    if(totalRequired == 3 and totalOptional > 1):
        df_game.loc[index,'hired'] = 1
        if(row['gender_pred'] ==0):
            df_game.loc[index, 'hired_bias'] =1
    
df_game
```
